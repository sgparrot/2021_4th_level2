**Attention(Q, K, V) = Attention Value**


  
어텐션 함수란, '쿼리(Query)'에 대해서 모든 '키(Key)'와의 유사도를 각각 구하고, 유사도를 각각의 '값(Value)'에 반영하는 것입니다. 

Attention value 어텐션 값 : 유사도가 반영된 '값(Value)'을 모두 더한 것

    Q = Query : t 시점의 디코더 셀에서의 은닉 상태

    K = Keys : 모든 시점의 인코더 셀의 은닉 상태들

    V = Values : 모든 시점의 인코더 셀의 은닉 상태들

## 2-1.  

````python
r = re.compile("c.p")   # .  1 
````

````python
print(r.search("cap"))
print(r.search("cup"))
print(r.search("accept")) # search ,      
print(r.search("cape"))
````

````python
print(r.search("cp"))
print(r.search("cab"))
````

````python
r = re.compile("pl?ay")   # ?   ,  
````

````python
print(r.search("pay"))
print(r.search("play"))
print(r.search("replay"))
print(r.search("player"))
````

````python
print(r.search("pray"))
````

````python
r = re.compile("go*d")   # *   0 
````

````python
print(r.search("gd"))
print(r.search("god"))
print(r.search("good"))
print(r.search("gooooooooood"))
````

````python
r = re.compile("go+d")   # +   1 (0   )
````

````python
print(r.search("god"))
print(r.search("good"))
print(r.search("gooooooooood"))
````

````python
print(r.search("gd"))
````

````python
r = re.compile("^ap")   # ^     
````

````python
print(r.search("ap"))
print(r.search("ape"))
print(r.search("apple"))
````

````python
print(r.search("cap"))
print(r.search("ale"))
````

````python
r = re.compile("ge{2}k")   # == geek
````

````python
print(r.search("geek"))
````

````python
print(r.search("gek"))
print(r.search("geeek"))
````

````python
r = re.compile("co{2,5}l")   # == cool, coool, cooool, coooool
````

````python
print(r.search("cool"))
print(r.search("coool"))
print(r.search("cooool"))
print(r.search("coooool"))
````

````python
print(r.search("col"))
print(r.search("cooooool"))
````

````python
### :      .

r = re.compile("co{2, 5}l")

print(r.search("cool"))
print(r.search("coool"))
print(r.search("cooool"))
print(r.search("coooool"))
````

````python
r = re.compile("co{5,}l")   # {n,} n , {0,} == *, {1,} == +
````

````python
print(r.search("cool"))
print(r.search("coool"))
print(r.search("cooool"))
print(r.search("coooool"))
print(r.search("cooooool"))
````

````python
r = re.compile("p[lr][aeo]y")   # []     1
````

````python
print(r.search("play"))
print(r.search("ploy"))
print(r.search("pray"))
print(r.search("prey"))
print(r.search("employ"))
````

````python
print(r.search("pay"))    #    
print(r.search("pLAy"))   #   
````

````python
r = re.compile("[a-zA-Z]")   #    
````

````python
print(r.search("a"))
print(r.search("A"))
print(r.search("aAbBcCdD"))
````

````python
r = re.compile("[^a-z]")   #         1 (^  )
````

````python
print(r.search("NumPy"))    #  N, P
print(r.search("123"))      # 
````

````python
print(r.search("print"))   #     None 
````

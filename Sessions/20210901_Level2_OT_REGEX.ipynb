# 2. Regular Expression (정규 표현식)

 정규표현식은 일종의 형식 언어로, string data를 처리하는 데에 유용하게 사용됩니다. Python뿐만 아니라 js, C, Perl 등 다양한 프로그래밍 언어에서 이를 지원하고 있고, 문법이 조금씩 상이하나 전체적인 틀은 큰 차이 없기 때문에 한 번 잘 익혀두면 다른 언어에서도 큰 어려움 없이 사용할 수 있습니다.

 모든 문법을 다루는 것이 현실적으로 불가능하기 때문에 자주 사용되는 것들 위주로 정리하였으며, 추가적으로 필요한 부분들은 직접 구글링 하여 사용하시는 것을 권장합니다. 저는 위에서 언급한 필수 자료와 더불어 다음과 같은 자료들을 참고하였습니다.

* 점프 투 파이썬 (https://wikidocs.net/1669)

````python
from pprint import pprint
import re   # python 내장 package
````

## 2-1. 각종 기호

````python
r = re.compile("c.p")   # .은 임의의 1개의 문자
````

````python
print(r.search("cap"))
print(r.search("cup"))
print(r.search("accept")) # search의 경우에는, 해당 문자열이 포함되면 더 길어도 괜찮음
print(r.search("cape"))
````

````python
print(r.search("cp"))
print(r.search("cab"))
````

````python
r = re.compile("pl?ay")   # ? 앞의 문자가 있을지, 없을지 모름
````

````python
print(r.search("pay"))
print(r.search("play"))
print(r.search("replay"))
print(r.search("player"))
````

````python
print(r.search("pray"))
````

````python
r = re.compile("go*d")   # * 앞의 문자가 0개 이상
````

````python
print(r.search("gd"))
print(r.search("god"))
print(r.search("good"))
print(r.search("gooooooooood"))
````

````python
r = re.compile("go+d")   # + 앞의 문자가 1개 이상(0개인 경우는 안 됨)
````

````python
print(r.search("god"))
print(r.search("good"))
print(r.search("gooooooooood"))
````

````python
print(r.search("gd"))
````

````python
r = re.compile("^ap")   # ^ 뒤에 있는 문자로 시작하는 것만
````

````python
print(r.search("ap"))
print(r.search("ape"))
print(r.search("apple"))
````

````python
print(r.search("cap"))
print(r.search("ale"))
````

````python
r = re.compile("ge{2}k")   # == geek
````

````python
print(r.search("geek"))
````

````python
print(r.search("gek"))
print(r.search("geeek"))
````

````python
r = re.compile("co{2,5}l")   # == cool, coool, cooool, coooool
````

````python
print(r.search("cool"))
print(r.search("coool"))
print(r.search("cooool"))
print(r.search("coooool"))
````

````python
print(r.search("col"))
print(r.search("cooooool"))
````

````python
### 주의: 다음과 같이 작성하면 정상적으로 작동하지 않는다.

r = re.compile("co{2, 5}l")

print(r.search("cool"))
print(r.search("coool"))
print(r.search("cooool"))
print(r.search("coooool"))
````

````python
r = re.compile("co{5,}l")   # {n,}는 n개 이상, {0,} == *, {1,} == +
````

````python
print(r.search("cool"))
print(r.search("coool"))
print(r.search("cooool"))
print(r.search("coooool"))
print(r.search("cooooool"))
````

````python
r = re.compile("p[lr][aeo]y")   # [] 안의 문자 중 아무거나 1개
````

````python
print(r.search("play"))
print(r.search("ploy"))
print(r.search("pray"))
print(r.search("prey"))
print(r.search("employ"))
````

````python
print(r.search("pay"))    # 비어있는 것은 안 됨
print(r.search("pLAy"))   # 당연히 대소문자 구별
````

````python
r = re.compile("[a-zA-Z]")   # 범위 지정도 가능 
````

````python
print(r.search("a"))
print(r.search("A"))
print(r.search("aAbBcCdD"))
````

````python
r = re.compile("[^a-z]")   # 괄호 안에 있는 것 외의 문자 중 아무거나 1개 (^와 다름에 주의)
````

````python
print(r.search("NumPy"))    # 대문자 N, P
print(r.search("123"))      # 숫자
````

````python
print(r.search("print"))   # 모두 소문자로 구성되어 있어 None 반환
````

## 2-2. 각종 함수

### 2-2-1. re.match()

지금까지는 re.search()를 사용했는데, 이는 "있기만 하면" matching을 시켜줍니다. 
반면, re.match()는 첫 글자부터 일치 여부를 확인합니다.

````python
r = re.compile("p[lr][aeo]y")
````

````python
print(r.match("play"))
print(r.match("ploy"))
print(r.match("pray"))
print(r.match("prey"))
print(r.match("player"))   # 뒤에 더 있는 건 못 잡아낸다
````

````python
print(r.match("employ"))
````

### 2-2-2. re.split()

````python
text = "I think therefore I am"

re.split(" ", text)
````

````python
text = "Alice, Bob, Carol, David, Eve"

re.split(", ", text)
````

### 2-2-3. re.findall()

````python
text = """
에스프레소: 2500원,
아메리카노: 2500원,
카푸치노: 3000원,
카페오레: 3000원,
카페모카: 3000원,
콜드브루: 3500원
바닐라라떼: 4500원,
샤케라또: 4500원

"""
re.findall("\d+원",text)   # \d는 [0-9]를 의미
````

### 2-2-4. re.sub()

sub는 substitute의 의미입니다. 즉, 이를 이용하여 특정 문자를 다른 문자로 대체할 수 있습니다. 아래의 글귀는 "MLP-Mixer: An all-MLP Architecture for Vision" (https://arxiv.org/pdf/2105.01601.pdf) 논문에서 발췌하였습니다.


````python
text = """
Convolutional Neural Networks (CNNs) are the go-to model for computer vision.
Recently, attention-based networks, such as the Vision Transformer, have also become popular.
In this paper we show that while convolutions and attention are both sufficient for good performance, neither of them are necessary.
We present MLP-Mixer, an architecture based exclusively on multi-layer perceptrons (MLPs).
"""
````

````python
re.sub("\W"," ",text)   # \W는 [^a-zA-Z0-9_]의 의미
````

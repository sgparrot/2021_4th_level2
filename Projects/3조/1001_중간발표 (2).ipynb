```python
def read_data(filename):

    with open(filename, "r", encoding="UTF8") as f:
        data = [line.split("\t") for line in f.read().splitlines()]

    return data
```

####데이터 정제 및 저장

```python
#vocabulary 작업

#토큰화된 데이터 flatten
words_list=[]

for i in words:
    for w in i:
        words_list.append(w)

print(words_list)

#word_to_index 작업

vocab={}

for w in words_list:
    if w not in vocab:
        vocab[w]=words_list.count(w)

print(vocab)
```

```python
import re
from tensorflow.python.keras.preprocessing.sequence import pad_sequences
```

```python
#Preprocessing 함수
def preprocessing(data, okt, remove_stopwords = False, stop_words = []):
    #숫자는 <n>으로 교체, 한글 및 공백을 제외한 문자 모두 제거.
    re.sub('[0-9]+', '<n>', data) 
    kor_text=re.sub("[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]", "", data)
    #okt 객체를 활용해 형태소 분리.
    morph_text=okt.morphs(kor_text, stem=True)
    #불용어 제거
    if remove_stopwords:
        clean_text=[token for token in morph_text if not token in stop_words]
    
    return clean_text
```

```python
#불용어 선정
stop_words=[]

#한국어 자연어 처리에서 흔히 쓰이는 불용어
stop_words_lib=[ '은', '는', '이', '가', '하', '아', '것', '들','의', '있', '되', '수', '보', '주', '등', '한', '에']

#voca에서 최대-최저 빈도 단어들
common_words=[]
for w in vocab:
    if vocab[w]>=1000:
        common_words.append(w)

few_words=[]
for w in vocab:
    if vocab[w]==1:
        few_words.append(w)

print(common_words)
print(few_words)
```

```python
#common_few_words의 품사에 따라 불용어 추가 여부 결정하기?
common_few_words=[]

print(tokenizer.pos(re.sub("[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]", "", str(common_words))))
print(tokenizer.pos(re.sub("[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]", "", str(few_words))))

stop_wrods=stop_words_lib+common_few_words
```

```python
clean_train_data = []

for data in X_train_data:
    # 비어있는 데이터에서 멈추지 않도록 string인 경우만 진행
    if type(data) == str:
        clean_train_data.append(preprocessing(data, tokenizer, remove_stopwords = True, stop_words=stop_words))
    else:
        clean_train_data.append([])  #string이 아니면 비어있는 값 추가

clean_test_data = []

for data in X_test_data:
    # 비어있는 데이터에서 멈추지 않도록 string인 경우만 진행
    if type(data) == str:
        clean_test_data.append(preprocessing(data, tokenizer, remove_stopwords = True, stop_words=stop_words))
    else:
        clean_test_data.append([])  #string이 아니면 비어있는 값 추가
```

```python
def write_txt(list, fname, sep) :
    file = open(fname, 'w')
    vstr = ''
    
    for a in list :
        vstr = vstr + str(a) + sep
    vstr = vstr.rstrip(sep)  # 마지막에도 추가되는  sep을 삭제 
    
    file.writelines(vstr)      # 한 라인씩 저장 
        
    file.close()
    print('[ 파일 저장 완료 ]')
```

```python
write_txt(clean_train_data, 'clean_train_data.txt', sep=',')
```

```python
write_txt(clean_test_data, 'clean_test_data.txt', sep=',')
```

```python
clean_vocab_sorted=sorted(clean_vocab.items(), key = lambda x:x[1], reverse=True)

word_to_index={}
i=0
for(word, freq) in clean_vocab_sorted:
    if freq>1:
        i=i+1
        word_to_index[word]=i

print(word_to_index)
```
